### Constructing a DataFrame using dict

```python
# Some input data - a small sample of the iris dataset
data = {'sepal_length': [6.9, 6.9, 4.8, 5.4, 4.6],
        'sepal_width': [3.2, 3.1, 3.4, 3.0, 3.6],
        'petal_length': [5.7, 5.1, 1.9, 4.5, 1.0],
        'petal_width': [2.3, 2.3, 0.2, 1.5, 0.2],
        'species': ['virginica', 'virginica', 'setosa', 'versicolor', 'setosa']}

df = pd.DataFrame(data)  #data를 dict의 형태로 만들어주고 그거를 입력값에 넣어주면 pandas안에 있는 data frame으로 만들어줌. 
df 
```

위의 코드에서, dict 안에 5개가 들어가 있음.
{ }안에

'sepal_lenth' 이런게 key가 되고 [ ]가 value가 됨 (key는 colum의 이름임)

개수가 맞아야함. key 부분에는 1개씩 들어가고 그리고 value 쪽에는 list의 형태로 들어가야함(5개)

### Constructing a DataFrame using pd.Series

data fame을 만드는 또 다른 방법
1. general한 csv 파일에서 pandas의 function, read_csv로 data frame 만듬 (제일 많이 쓰는 방법)
-> excel 파일의 형태로 주고받고 저장하는 게 더 합리적임.
2. dict 파일을 통으로 만들어서 data frame을 만듬
3. series를 써서 column의 내용 부분을 따로 만들고 결국은 data frame을 dict를 이용해서 만듬

```python
sepal_length = pd.Series([6.9, 6.9, 4.8, 5.4, 4.6])  #한 column에 해당되는 내용을 series라는 함수를 이용해서 따로 만듬.
sepal_width = pd.Series([3.2, 3.1, 3.4, 3.0, 3.6])
petal_length = pd.Series([5.7, 5.1, 1.9, 4.5, 1.0])
petal_width = pd.Series([2.3, 2.3, 0.2, 1.5, 0.2])
species = pd.Series(['virginica', 'virginica', 'setosa', 'versicolor', 'setosa'])

df = pd.DataFrame({'sepal_length': sepal_length, 'sepal_width': sepal_width, 'petal_length': petal_length, 'petal_width': petal_width, 'SPECIES': species})
#dict를 여기 위에서 만들어주는 거임.
df
```

### Changing column names

```python
df.columns  #위의 column 정보만 나옴.
```

```python
# Change the columns (species 부분을 대문자에서 소문자로 바꾸고 싶다)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```

### Accessing columns

```python
# Access one column 
df.loc[:, 'sepal_length']  #loc는 location, comma 왼쪽의 부분이 row(행)이고 오른쪽 부분이 column(열)
```

```python
# Access one column (short-hand)
df['sepal_length']
```

```python
# Access one column (shortest-hand)
df.sepal_length
```

```python
# Access two columns
df.loc[:, ['sepal_length', 'sepal_width']]
```

```python
# Access two columns (short-hand)
df[['sepal_length', 'sepal_width']]
```

### Accessing rows

```python
# Access one row 
df.loc[3, :]  #결과값이 세로로 되어있지만, column이 아니고 row임
```

```python
# Access two rows
df.loc[[3, 0], :]
```

```python
# Access a range of rows
df.loc[2:5, :] #2:5는 2에서 5를 빼고까지 불러와라
```

```python
# Filter keeping only rows with `sepal_length` > 5
df[df.sepal_length > 5]
```

### Accessing a cell

```python
df.loc[:, 'sepal_length'][4] #모든 행에서 'sepal_length'에 해당하는 것 중에서 거기서 4번째꺼.
```

```python
df['sepal_length'][4]
```

```python
df.sepal_length[4]
```

column access에서 봤듯이 다양하게 줄여서 쓸 수 있음.

### Adding/Removing new columns

```python
df['new_column'] = 1  #column access할때 적는 방법임, df['new_column']만 쓰면 에러 메세지 나옴.
df
```

```python
df['new_column'] = [1, 2, 3, 4, 5]
df
```

```python
df = df.drop('new_column', axis=1) # axis=1: column  #new_column에 해당하는 것을 drop해라.
#new_column이 row인지 column에 있는 정보인지 모름. axis=1이라고 적고 있으면 column에서 axis=0이라고 적혀 있으면 세로에서 찾고 있을거임
#axis=0이 행 정보이고 axis=1은 열 정보임.
df
```

위의 코드에서, df = 이라고 써줘야 업데이트가 됨.

df.drop만 쓰면 업데이트안됨.

### Adding/Removing new rows

```python
df.loc[5, :] = ''   #:은 all column, 그리고 5는 row 부분에 추가해라. ' '는 string. ' '이걸 써줘야 에러가 안남
df
```

```python
df = df.drop(5, axis=0) # axis=0: row
df
```
