### [Pytorch Hub](https://pytorch.org/hub/research-models)

들어가보면 pytorch hub for researchers (연구자들이 많이 쓰는거임)

> image generation (e.g. [PGAN](https://pytorch.org/hub/facebookresearch_pytorch-gan-zoo_pgan/))

```python
import torch
model = torch.hub.load('facebookresearch/pytorch_GAN_zoo:hub', 'PGAN', model_name='celebAHQ-512', pretrained=True)
#말 그대로 AI가 예술가의 활동을 하는 거임. 그림도 그리고 사진도 찍는데, 기존의 것이 아니라 새로운 것을 가져오는거임.
#PGAN은 알고리즘이라고 생각하면 됨.
```

```python
num_images = 4  #이미지 개수 4개
noise, _ = model.buildNoiseData(num_images)  #4개 넣으면 noise 4개 만들어주고
with torch.no_grad():
    generated_images = model.test(noise)   #noise를 넣으면 이미지가 만들어짐. noise를 만들어주는 이유? - 원래부터 모델이 그렇게 훈련되어져 있음. random한 숫자들이 random한 noise들을 만들어내는거임.

# let's plot these images using torchvision and matplotlib   #matplotlib을 이용해서 visualize해주는거임
import matplotlib.pyplot as plt
import torchvision
grid = torchvision.utils.make_grid(generated_images.clamp(min=-1, max=1), scale_each=True, normalize=True)
plt.imshow(grid.permute(1, 2, 0).cpu().numpy())
```
