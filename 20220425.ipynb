# Auto Correlation

Auto-Correlation은 자기 상관이라고도 불리며, 동일한 시간구간으로 나누어진 벡터로 구성된 시계열 자료에서, 현재의 개체수가 이전의 개체수와 어떻게 연관이 되어 있는가를 나타낸 것이다. 
- correlation은 두 변수 간의 선형적 관계를 측정하고자 할 때 사용한다.
- 시간 또는 공간적으로 연속된 관측치들 간에 존재하는 상관관계로서 시계열 데이터에 내제하는 시점간의 상관이다.
- 시간항에 대해서 shifted된 자기 자신에 대해서 correlation한것이라 auto correlation이라고 불리는 것이다.
-corrcoef이 +1.0이라면 완전히 동일한 신호로 corrcoef가 -1.0이라는 완전히 반대인 신호라고 생각하면 된다. 
- 월별 자료인 경우, 이번 달 개체수가 지난 달 개체수와 어떻게 연관성이 있는가를 나타낸 것이 1차 자기상관이며, 이번 달 개체수가 k 시간차의 개체수와 어떻게 연관성이 있는가를 나타내는 것이 k차 자기 상관이다.
- 부분 자기상관은 시계열 자료에서 현재의 개체수는 다른 시간 구간의 개체수와 상관관계를 통제하고, 시간차 k개의 개체수와 어떻게 연관이 되어 있는가를 나타낸 것이다. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
s = pd.Series(np.sin(range(1,100))) + pd. Series(np.random.randn(99))

plt.figure(facecolor='w', figsize = (20,3), dpi=120)
plt.plot(s)
plt.show()
```

```python
s.autocorr(lag=1)
```

# Root Mean Square

Root Mean Squar(RMS) 
- 실효치, 이차평균(quadratic mean)이라고도 한다.
- 특정 값이 음과 양을 오갈때 사용
- 각 값들의 제곱에 대한 평균을 낸 다음 그 값에 제곱근을 취해 계산한다.

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

X = 4 * np.random.randn(1000,1)
X_b = np.c_[np.ones((1000,1)), X]

Y = 2 + 3 * X + np.random.randn(1000,1)

plt.plot(X,Y, '.')
plt.xlim(0,4)
plt.ylim(0,15)

plt.xlabel(r'x', fontsize = 8)
plt.ylabel(r'y', fontsize=8)
plt.title('How to calculate the mean squared error in python', fontsize=8)
```

```python
theta = np.array([[-1.4], [5.0]])

X_new = np.array([[0],[4]])
X_new_b = np.c_[np.ones((2,1)), X_new]

plt.plot(X_new, X_new_b.dot(theta), '-')
plt.xlim(0,4)
plt.ylim(0,15)

plt.xlabel(r'x', fontsize=8)
plt.ylabel(r'y', fontsize=8)
plt.title('How to calculate the mean squared error in python', fontsize=8)
```

```python
plt.close()
```

```python
Y_predict = X_b.dot(theta)
print(Y_predict.shape, X_b.shape, theta.shape)
mse = np.sum((Y_predict-Y)**2) / 1000.0
print('mse: ', mse)
```
